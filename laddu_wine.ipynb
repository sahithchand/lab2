are large.

## K-Means

```python
KM = KMeans(init="random",n_clusters=3,n_init=10,max_iter=300,random_state=42)
res = KM.fit(wine)

label_pred_KM = KM.labels_
print("Labels predicted by K-Means:", label_pred_KM)
print('Length of labels is same as data entry', label_pred_KM.shape)

centroids_KM = KM.cluster_centers_
print("Shape of centroids_KM", centroids_KM.shape)
print("Centroids:", centroids_KM)

inertia_KM = KM.inertia_
print("Inertia:", inertia_KM)
```

We can plot the pairwise relationships of `winenorm` with the colored label predicted bt K-Means method.

```python
# Pairplot
wine["cluster"] = label_pred_KM
sns_plot = sns.pairplot(wine)
```

```python
# Filter the dataset to include only rows with cluster 0
cluster1_data = wine[wine['cluster'] == 0]

# Select the alcohol-related features
alcohol_features = ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280', 'Proline']

# Calculate the correlation matrix
correlation_matrix = cluster1_data[alcohol_features].corr()

plt.figure(figsize=(9, 9))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap for Alcohol-Related Features in Cluster 0')
plt.show()
```
