```python
classification_counts = df['Classification'].value_counts()

# Step 3: Create the plot
plt.figure(figsize=(15, 13))
sns.barplot(x=classification_counts.values, y=classification_counts.index, palette='husl')
plt.title('Jobs and number of Postings')
plt.xlabel('Number of Postings')
plt.ylabel('Classification')
plt.tight_layout()

# Show plot
plt.show()
```

```python
a=df.query('LowestSalary==0.0')
b=df.query('LowestSalary==30.0')
c=df.query('LowestSalary==40.0')
d=df.query('LowestSalary==50.0')
df1=pd.DataFrame({'Lowest Salary':[0.0,30.0,40.0,50.0]})
df1['Id']=[a['Id'].value_counts().sum(),b['Id'].value_counts().sum(),c['Id'].value_counts().sum(),d['Id'].value_counts().sum()]
df1['Title']=[a['Title'].value_counts().sum(),b['Title'].value_counts().sum(),c['Title'].value_counts().sum(),d['Title'].value_counts().sum()]
df1['Company']=[a['Company'].value_counts().sum(),b['Company'].value_counts().sum(),c['Company'].value_counts().sum(),d['Company'].value_counts().sum()]
df1['Date']=[a['Date'].value_counts().sum(),b['Date'].value_counts().sum(),c['Date'].value_counts().sum(),d['Date'].value_counts().sum()]
df1['Location']=[a['Location'].value_counts().sum(),b['Location'].value_counts().sum(),c['Location'].value_counts().sum(),d['Location'].value_counts().sum()]
df1['Area']=[a['Area'].value_counts().sum(),b['Area'].value_counts().sum(),c['Area'].value_counts().sum(),d['Area'].value_counts().sum()]
df1['Classification']=[a['Classification'].value_counts().sum(),b['Classification'].value_counts().sum(),c['Classification'].value_counts().sum(),d['Classification'].value_counts().sum()]
df1['SubClassification']=[a['SubClassification'].value_counts().sum(),b['SubClassification'].value_counts().sum(),c['SubClassification'].value_counts().sum(),d['SubClassification'].value_counts().sum()]
df1['Requirement']=[a['Requirement'].value_counts().sum(),b['Requirement'].value_counts().sum(),c['Requirement'].value_counts().sum(),d['Requirement'].value_counts().sum()]
df1['FullDescription']=[a['FullDescription'].value_counts().sum(),b['FullDescription'].value_counts().sum(),c['FullDescription'].value_counts().sum(),d['FullDescription'].value_counts().sum()]
df1['HighestSalary']=[a['HighestSalary'].value_counts().sum(),b['HighestSalary'].value_counts().sum(),c['HighestSalary'].value_counts().sum(),d['HighestSalary'].value_counts().sum()]
df1['JobType']=[a['JobType'].value_counts().sum(),b['JobType'].value_counts().sum(),c['JobType'].value_counts().sum(),d['JobType'].value_counts().sum()]
df1['Label']=[0.0,30.0,40.0,50.0]
df1['Count']=[a['Id'].value_counts().sum(),b['Id'].value_counts().sum(),c['Id'].value_counts().sum(),d['Id'].value_counts().sum()]
print(df1)
```

```python
import matplotlib.pyplot as plt

# Sample data
lowest_salary_counts = {
    0.0: 3894,
    30.0: 1842,
    40.0: 2519,
    50.0: 1844
}

# Extracting data for plotting
labels = lowest_salary_counts.keys()
sizes = lowest_salary_counts.values()

# Creating donut chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))

# Draw a circle at the center of pie to make it a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  
plt.title('Job posts by salary range')
plt.show()
```
